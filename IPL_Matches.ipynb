```python
match_data_df = pd.read_csv('https://raw.githubusercontent.com/insaid2018/Term-1/master/Data/Projects/deliveries.csv')
match_data_df.head() #using one more dataset deliveries.csv 
```

```python
team_score = match_data_df.groupby(['match_id', 'inning'])['total_runs'].sum().unstack().reset_index()
team_score.columns = ['match_id', 'Team1_score', 'Team2_score', 'Team1_superover_score', 'Team2_superover_score']
match_data_merge = pd.merge(csv_df, team_score, left_on = 'id', right_on = 'match_id', how = 'outer')
```

```python
cols = ['match_id', 'season','city','date','team1','team2', 'toss_winner', 
        'toss_decision', 'result', 'dl_applied', 'winner', 'Team1_score','Team2_score', 
        'win_by_runs', 'win_by_wickets', 'Team1_superover_score', 'Team2_superover_score',
        'player_of_match', 'venue', 'umpire1', 'umpire2', 'umpire3']
match_data_merge = match_data_merge[cols]
match_data_merge.head(2)
```
